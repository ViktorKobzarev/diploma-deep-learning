# вычисляем AUC

```python
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Вычисляем AUC для каждого класса
def calculate_class_auc(probabilities_meta, labels):
    """
    Вычисляет AUC для каждого класса отдельно
    """
    n_classes = probabilities_meta.shape[1]
    auc_per_class = []
    
    print("=== AUC по каждому классу ===")
    for i in range(n_classes):
        # Получаем вероятности и метки для текущего класса
        class_probs = probabilities_meta[:, i]
        class_labels = labels[:, i]
        
        # Проверяем, есть ли оба класса (0 и 1) в метках
        if len(np.unique(class_labels)) < 2:
            print(f"Класс {i}: AUC = N/A (все метки одинаковые)")
            auc_per_class.append(np.nan)
            continue
        
        try:
            # Вычисляем AUC для текущего класса
            auc = roc_auc_score(class_labels, class_probs)
            auc_per_class.append(auc)
            print(f"Класс {i}: AUC = {auc:.4f}")
        except Exception as e:
            print(f"Класс {i}: Ошибка при расчете AUC - {e}")
            auc_per_class.append(np.nan)
    
    return np.array(auc_per_class)

# Вычисляем общий AUC (macro-average и micro-average)
def calculate_overall_auc(probabilities_meta, labels):
    """
    Вычисляет общие метрики AUC
    """
    print("\n=== Общие метрики AUC ===")
    
    # 1. Macro-average AUC (среднее по классам)
    auc_per_class = calculate_class_auc(probabilities_meta, labels)
    valid_auc = auc_per_class[~np.isnan(auc_per_class)]
    
    if len(valid_auc) > 0:
        macro_auc = np.mean(valid_auc)
        print(f"\nMacro-average AUC: {macro_auc:.4f}")
    else:
        macro_auc = np.nan
        print("\nMacro-average AUC: N/A (нет валидных классов)")
    
    # 2. Micro-average AUC (объединяем все предсказания и метки)
    # Flatten все предсказания и метки
    micro_probs = probabilities_meta.ravel()
    micro_labels = labels.ravel()
    
    if len(np.unique(micro_labels)) >= 2:
        micro_auc = roc_auc_score(micro_labels, micro_probs)
        print(f"Micro-average AUC: {micro_auc:.4f}")
    else:
        micro_auc = np.nan
        print(f"Micro-average AUC: N/A (все метки одинаковые)")
    
    # 3. AUC по взвешенным классам (weighted average)
    if len(valid_auc) > 0:
        # Веса = доля положительных примеров в каждом классе
        class_weights = []
        valid_indices = np.where(~np.isnan(auc_per_class))[0]
        
        for idx in valid_indices:
            class_labels = labels[:, idx]
            weight = np.mean(class_labels)  # доля положительных примеров
            class_weights.append(weight)
        
        class_weights = np.array(class_weights)
        # Нормализуем веса
        if class_weights.sum() > 0:
            class_weights = class_weights / class_weights.sum()
            weighted_auc = np.sum(valid_auc * class_weights)
            print(f"Weighted-average AUC: {weighted_auc:.4f}")
        else:
            weighted_auc = np.nan
            print(f"Weighted-average AUC: N/A (все веса равны 0)")
    else:
        weighted_auc = np.nan
    
    return {
        'auc_per_class': auc_per_class,
        'macro_auc': macro_auc,
        'micro_auc': micro_auc,
        'weighted_auc': weighted_auc
    }

# Функция для визуализации ROC кривых
def plot_roc_curves_all(probabilities_meta, labels, auc_results):
    """
    Строит ROC кривые для каждого класса
    """
    n_classes = probabilities_meta.shape[1]
    
    # Определяем размер сетки для графиков
    n_cols = 4
    n_rows = int(np.ceil(n_classes / n_cols))
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3.5))
    axes = axes.flatten() if n_classes > 1 else [axes]
    
    for i in range(n_classes):
        class_probs = probabilities_meta[:, i]
        class_labels = labels[:, i]
        
        # Проверяем, есть ли оба класса
        if len(np.unique(class_labels)) < 2:
            axes[i].text(0.5, 0.5, f'Class {i}\nOnly one class',
                        ha='center', va='center', fontsize=12)
            axes[i].set_title(f'Class {i}: AUC = N/A')
            axes[i].set_xlim([0, 1])
            axes[i].set_ylim([0, 1])
        else:
            # Вычисляем ROC кривую
            fpr, tpr, thresholds = roc_curve(class_labels, class_probs)
            auc_value = auc_results['auc_per_class'][i]
            
            # Строим ROC кривую
            axes[i].plot(fpr, tpr, 'b-', linewidth=2, label=f'AUC = {auc_value:.3f}')
            axes[i].plot([0, 1], [0, 1], 'r--', linewidth=1, label='Random')
            axes[i].set_title(f'Class {i}: AUC = {auc_value:.3f}')
            axes[i].set_xlabel('False Positive Rate')
            axes[i].set_ylabel('True Positive Rate')
            axes[i].legend(loc='lower right')
            axes[i].grid(True, alpha=0.3)
    
    # Скрываем лишние субплотов если есть
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.suptitle('ROC Curves for Each Class', fontsize=16, y=1.02)
    plt.tight_layout()
    plt.show()
    
    # Также строим общий график со всеми кривыми
    plt.figure(figsize=(10, 8))
    
    for i in range(n_classes):
        class_probs = probabilities_meta[:, i]
        class_labels = labels[:, i]
        
        if len(np.unique(class_labels)) >= 2:
            fpr, tpr, _ = roc_curve(class_labels, class_probs)
            auc_value = auc_results['auc_per_class'][i]
            plt.plot(fpr, tpr, lw=1.5, alpha=0.7, label=f'Class {i}: AUC = {auc_value:.3f}')
    
    plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random (AUC = 0.5)')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('All ROC Curves Together', fontsize=14)
    plt.legend(loc='lower right', fontsize=9)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Основная функция для расчета всех метрик
def calculate_all_auc_metrics(probabilities_meta, labels):
    """
    Основная функция для расчета всех AUC метрик
    """
    # Проверяем размерности
    assert probabilities_meta.shape == labels.shape, \
        f"Размерности не совпадают: probabilities_meta {probabilities_meta.shape}, labels {labels.shape}"
    
    print(f"Размерность probabilities_meta: {probabilities_meta.shape}")
    print(f"Размерность labels: {labels.shape}")
    print(f"Количество классов: {probabilities_meta.shape[1]}")
    
    # Вычисляем все AUC метрики
    auc_results = calculate_overall_auc(probabilities_meta, labels)
    
    # Выводим статистику по меткам
    print("\n=== Статистика по меткам ===")
    n_classes = labels.shape[1]
    for i in range(n_classes):
        class_labels = labels[:, i]
        pos_count = np.sum(class_labels == 1)
        neg_count = np.sum(class_labels == 0)
        pos_ratio = pos_count / len(class_labels)
        print(f"Класс {i}: {pos_count} положительных, {neg_count} отрицательных ({pos_ratio:.2%})")
    
    # Визуализируем ROC кривые
    plot_roc_curves_all(probabilities_meta, labels, auc_results)
    
    return auc_results
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Расчет метрик для каждого класса
n_classes = labels.shape[1]
print("Метрики по каждому классу:")
print("-" * 60)

for i in range(n_classes):
    class_pred = binary_predictions[:, i]
    class_true = labels[:, i]
    
    # Если в классе нет положительных примеров
    if np.sum(class_true) == 0:
        print(f"{diseases[i]}: Нет положительных примеров в метках")
        continue
    
    # Метрики для текущего класса
    precision = precision_score(class_true, class_pred, zero_division=0)
    recall = recall_score(class_true, class_pred, zero_division=0)
    f1 = f1_score(class_true, class_pred, zero_division=0)
    
    print(f"{diseases[i]}:")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1-score:  {f1:.4f}")

print("\nОбщие метрики:")
print("-" * 60)

# Micro-average
precision_micro = precision_score(labels, binary_predictions, average='micro', zero_division=0)
recall_micro = recall_score(labels, binary_predictions, average='micro', zero_division=0)
f1_micro = f1_score(labels, binary_predictions, average='micro', zero_division=0)

print(f"Micro-average:")
print(f"  Precision: {precision_micro:.4f}")
print(f"  Recall:    {recall_micro:.4f}")
print(f"  F1-score:  {f1_micro:.4f}")

# Macro-average
precision_macro = precision_score(labels, binary_predictions, average='macro', zero_division=0)
recall_macro = recall_score(labels, binary_predictions, average='macro', zero_division=0)
f1_macro = f1_score(labels, binary_predictions, average='macro', zero_division=0)

print(f"\nMacro-average:")
print(f"  Precision: {precision_macro:.4f}")
print(f"  Recall:    {recall_macro:.4f}")
print(f"  F1-score:  {f1_macro:.4f}")

# Weighted-average
precision_weighted = precision_score(labels, binary_predictions, average='weighted', zero_division=0)
recall_weighted = recall_score(labels, binary_predictions, average='weighted', zero_division=0)
f1_weighted = f1_score(labels, binary_predictions, average='weighted', zero_division=0)

print(f"\nWeighted-average:")
print(f"  Precision: {precision_weighted:.4f}")
print(f"  Recall:    {recall_weighted:.4f}")
print(f"  F1-score:  {f1_weighted:.4f}")
```
